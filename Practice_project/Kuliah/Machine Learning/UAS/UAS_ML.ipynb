## Import Modul yang diperlukan

```python
import pandas as pd
import numpy as np
```

## Membaca dan melihat dataset

```python
data = pd.read_excel('Harga Pangan.xlsx')
data.head()
```

## Menghapus kolom yang tidak diperlukan

```python
data.drop('No', axis=1, inplace=True)
data.head()
```

## Mengubah struktur dataframe

### Transpose dataframe agar tanggal menjadi baris dan komoditas menjadi kolom

```python
data_transposed = data.T
data_transposed.head()
```

### Mereset indeks setelah proses transpose

```python
data_transposed.reset_index(inplace=True)
data_transposed.head()
```

### Memberikan nama pada kolum (index,0,1,2,3,4,5) agar menjadi baris pertama

```python
data_transposed.columns = data_transposed.iloc[0]
data_transposed.head()
```

### Menghapus baris pertama

```python
data_transposed = data_transposed.drop(0)
data_transposed.head()
```

### Reset indeks agar dimulai dari 0 kembali

```python
data_transposed.reset_index(drop=True, inplace=True)
```

### Mengubah kolum 'Komoditas (Rp)' menjadi 'Tanggal'

```python
data_transposed = data_transposed.rename(columns={'Komoditas (Rp)' : 'Tanggal'})
data_transposed.head()
```

```python
data_transposed.info()
```

## Menyeleksi kolum yang akan digunakan

```python
# Karena pada tugas diperlukan 1 kategori
# Maka disini akan memilih satu kategori
# Diantara 7 kolom yaitu Beras
df = data_transposed.drop(['Beras Kualitas Medium I',
                           'Daging Ayam','Daging Ayam Ras Segar',
                           'Daging Sapi',
                           'Daging Sapi Kualitas 1'],
                          axis=1)
df.head()
```

^ Di atas ada data yang nilainya kosong yaitu '-' ini perlu diubah ^

## Mengatasi Missing Value

```python
# Mengatasi nilai kosong
df = df.replace('-', 0)
print(df.info())
df.head()
```

^ Di atas Nilai yang '-' Di ubah menjadi 0 agar kolum Beras dapat di ubah ke tipe data numerik, dimana awalnya adalah object ^

```python
# Mengubah tipe data kolum "Beras" dari object menjadi float
# Dan disini yang nilainya 0 akan menjadi null
df['Beras'] = df['Beras'].str.replace(',','').astype(float)
print(df.info())
df.head()
```

^ Jika dilihat pada info() kolum 'Beras' sudah berubah menjadi float ^

```python
# Mengecek Missing Value pada data
df.isna().sum()
```

^ Di atas terdapat 10 baris pada kolom Beras yang memiliki nilai null ^

```python
# Mengatasi missing value dengan mengisi nilai null
# Menggunakan rata-rata kolum beras
mean_harga_beras = int(df['Beras'].mean())
df_non_null = df.fillna(mean_harga_beras)
print('Nilai Null Pada Kolum Beras :', df_non_null['Beras'].isna().sum())
df_non_null
```

## Mengubah tipe data 'Tanggal'

```python
# Mengubah dtype Tanggal dari object ke datetime
df_clean = df_non_null

# Menghapus spasi yang ada pada tanggal and konversi "Tanggal" ke datetime
df_clean["Tanggal"] = pd.to_datetime(df_clean["Tanggal"].str.replace(' ', ''), format="%d/%m/%Y")
```

```python
print(df_clean.info())
df_clean.head()
```

## Ekstrasi dan data splitting

```python
# Ekstrasi fitur dari kolum 'Tanggal'
df_clean["Year"] = df_clean["Tanggal"].dt.year
df_clean["Month"] = df_clean["Tanggal"].dt.month
df_clean["Day"] = df_clean["Tanggal"].dt.day
df_clean["DayOfYear"] = df_clean["Tanggal"].dt.dayofyear
df_clean.head()
```

```python
# Assign variabel Fitur dan Target
X = df_clean[["Year", "Month", "Day", "DayOfYear"]]
y = df_clean["Beras"]
```

```python
# Split data training dan testing
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

## Model Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
# Pemodelan dengan naiive bayes
model = GaussianNB()
model.fit(X_train, y_train)
```

## Evaluasi Model

```python
# Evaluasi model
prediksi = model.predict(X_test)
accuracy = (prediksi == y_test).mean()
print(f"Accuracy: {accuracy * 100:.2f}%")
```

### Prediksi harga beras 30 hari ke depan

```python
# Membuat data 30 hari ke depan
import numpy as np
tanggal_kedepan = pd.date_range(start=df_clean["Tanggal"].max(), periods=30, freq='D')
tanggal_kedepan
```

```python
# Membuat fitur data 30 hari ke depan
data_kedepan = pd.DataFrame({
    "Tanggal": tanggal_kedepan,
    "Year": tanggal_kedepan.year,
    "Month": tanggal_kedepan.month,
    "Day": tanggal_kedepan.day,
    "DayOfYear": tanggal_kedepan.dayofyear
})
data_kedepan.head()
```

```python
# Prediksi harga beras 30 hari ke depan
prediksi_kedepan = model.predict(data_kedepan[["Year", "Month", "Day", "DayOfYear"]])
prediksi_kedepan
```

```python
# Merge prediksi harga beras dan fitur
data_kedepan["Predicted_Beras"] = prediksi_kedepan
data_kedepan
```

```python
# Konversi "Tanggal"
data_kedepan["Tanggal"] = data_kedepan["Tanggal"].dt.strftime("%d/%m/%Y")
```

```python
data_kedepan
```

## Visualisasi Data Prediksi

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Melt DataFrame untuk merubah data menjadi format long
df_future = data_kedepan.melt(id_vars=['Tanggal'], value_vars=['Predicted_Beras'],
                     var_name='Komoditas', value_name='Harga')


# Plot menggunakan Seaborn
plt.figure(figsize=(15, 6))
sns.set_style("dark")
ax = sns.lineplot(
    data=df_future, 
    x='Tanggal', 
    y='Harga', 
    hue='Komoditas',
    palette="rocket"
)
plt.xlabel('Tanggal')
plt.ylabel('Harga')
plt.title('Visualisasi Time Series Harga Beras Setelah di Prediksi')
plt.legend(title='Komoditas')
plt.grid(True)
ax.set_facecolor('white')
ax.figure.set_facecolor('white')
plt.show()
```
